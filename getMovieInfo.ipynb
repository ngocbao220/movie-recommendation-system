```python
import requests
import time
import os
import webbrowser

API_KEY = "ff48b02cdcd1f6e40df93cb3ff292031"  # ‚ùó n√™n d√πng bi·∫øn m√¥i tr∆∞·ªùng
TMDB_ID = 862

# ===== G·ªåI TMDB: chi ti·∫øt phim =====
movie_url = f"https://api.themoviedb.org/3/movie/{TMDB_ID}"
params = {"api_key": API_KEY}

start = time.time()
res_movie = requests.get(movie_url, params=params)
elapsed = time.time() - start

if res_movie.status_code != 200:
    print("‚ùå L·ªói khi l·∫•y th√¥ng tin phim")
    exit()

movie_data = res_movie.json()
movie_title = movie_data.get("title", "Unknown Title")
print(f"üé¨ T√™n phim: {movie_title}")
print(f"‚è±Ô∏è Th·ªùi gian g·ªçi API chi ti·∫øt phim: {elapsed:.3f} gi√¢y")

# ===== G·ªåI TMDB: danh s√°ch video =====
video_url = f"https://api.themoviedb.org/3/movie/{TMDB_ID}/videos"
res_video = requests.get(video_url, params=params)
data = res_video.json()

youtube_key = None
for v in data["results"]:
    if v["site"] == "YouTube" and v["type"] == "Trailer":
        youtube_key = v["key"]
        break

if not youtube_key:
    print("‚ùå Kh√¥ng t√¨m th·∫•y trailer")
    exit()

print(f"üéûÔ∏è YouTube key: {youtube_key}")

# ===== T·∫†O FILE HTML STREAM =====
html_content = f"""
<!DOCTYPE html>
<html>
<head>
    <meta charset="utf-8">
    <title>{movie_title} - Trailer</title>
    <style>
        html, body {{
            margin: 0;
            padding: 0;
            width: 100%;
            height: 100%;
            background: black;
        }}
        iframe {{
            width: 100%;
            height: 100%;
            border: none;
        }}
    </style>
</head>
<body>
    <iframe
        src="https://www.youtube.com/watch?v={youtube_key}"
        allow="autoplay; encrypted-media"
        allowfullscreen>
    </iframe>
</body>
</html>
"""

output_file = "trailer.html"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(html_content)

print(f"‚úÖ ƒê√£ t·∫°o file stream: {os.path.abspath(output_file)}")
print(f"https://www.youtube.com/watch?v={youtube_key}")

# M·ªü tr·ª±c ti·∫øp
webbrowser.open(f"file://{os.path.abspath(output_file)}")
```
